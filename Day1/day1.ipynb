```python
file = open("input.txt", "r")
content = file.read()

split_lines = content.splitlines()
# split_lines = "two1nine\neightwothree\nabcone2threexyz\nxtwone3four\n4nineeightseven2\nzoneight234\n7pqrstsixteen".splitlines()
```

```python
answer = 0
for s in split_lines:

    left, right = 0, len(s) - 1

    # move left
    while not s[left].isnumeric():
        left+=1
    
    while not s[right].isnumeric():
        right-=1
    
    answer+=int(s[left]) * 10
    answer+=int(s[right])

print(answer)
        
```
